```python
todo_list=[]

def Add_list():
    item = input("enter your task to add in list")
    todo_list.append(item)

def Display_list():
    print("-----------------")
    print("TO Do LIST:")
    print("-----------------")
    for index, item in enumerate(todo_list,start=1):
        print(f"{index} - {item}")
def Remove_list():
    Display_list()
    index= int(input("Enter your number to remove"))-1
    if 0 <= index <len(todo_list):
        removed_item = todo_list.pop(index)
        print("removed item is: ", removed_item)
    else:
        print("invalid")

print("##################################")
print("To Do List App")
print("##################################")
print("1 - Add To List")
print("2 - Display The List")
print("3 - Remove From List")
print("4 - Exit")

while True:
    option = input("Enter your option:")

    if option == '1':
        Add_list()
    elif option == '2':
        Display_list()
    elif option == '3':
        Remove_list()
    elif option == '4':
        print("Exit")
        break
    else:
        print("Invalid Option")
```
